# Introduction
This data set contains information on hospitals in Kenya. The data set contains data from 12,394 hospitals and the Keph level of each hospital. This information will be used to rate each hospital and hence we will be able to recommend the best hospital to visit in an area based on the service the patient requires.

# Data Leakage

```python
df_a = df_hospitals # encoding without target
```

```python
Keph_level = pd.get_dummies(df_a['Keph_level'])
print(Keph_level)
```

```python
Keph_level.value_counts()
```

```python
y_a = df_a['Keph_level']
x_a = df_a.drop(columns=['Keph_level'])
x_a.head(3)
```

```python
# Train test split to avoid data leakage

from sklearn.model_selection import train_test_split

x_a_train, x_a_test, y_a_train, y_a_test = train_test_split(x_a, y_a, test_size=0.3, random_state=1)
```

```python
#Encoding WITHOUT target (using df_a)
#Label Encoder
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
le = LabelEncoder()
Open_whole_day_le = le.fit_transform(x_a_train.Open_whole_day)

#Apply to x_train and x_test

x_a_train.loc[:, 'Open_whole_day'] = le.fit_transform(x_a_train.loc[:, 'Open_whole_day'])
x_a_train.loc[:, 'Open_public_holidays'] = le.fit_transform(x_a_train.loc[:, 'Open_public_holidays'])
x_a_train.loc[:, 'Open_weekends'] = le.fit_transform(x_a_train.loc[:, 'Open_weekends'])
x_a_train.loc[:, 'Open_late_night'] = le.fit_transform(x_a_train.loc[:, 'Open_late_night'])

x_a_test.loc[:, 'Open_whole_day'] = le.fit_transform(x_a_test.loc[:, 'Open_whole_day'])
x_a_test.loc[:, 'Open_public_holidays'] = le.fit_transform(x_a_test.loc[:, 'Open_public_holidays'])
x_a_test.loc[:, 'Open_weekends'] = le.fit_transform(x_a_test.loc[:, 'Open_weekends'])
x_a_test.loc[:, 'Open_late_night'] = le.fit_transform(x_a_test.loc[:, 'Open_late_night'])
```

```python
#OneHotEncoder

from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder(sparse=False, drop='first', handle_unknown='error')#

#Apply for x_train set

nominal = onehot.fit_transform(x_a_train[['Facility_type_category', 'Owner_type', 'Regulatory_body']])
df_nominal = pd.DataFrame(index=x_a_train.index, data=nominal, columns=onehot.get_feature_names(['Facility_type_category', 'Owner_type', 'Regulatory_body']))
x_a_train = pd.concat([x_a_train, df_nominal], axis=1)
x_a_train.head().T
```

```python
x_a_train.drop(columns=['Facility_type_category', 'Owner_type', 'Regulatory_body'], inplace=True)
x_a_train.head().T
```

```python
nominal_test = onehot.fit_transform(x_a_test[['Facility_type_category', 'Owner_type', 'Regulatory_body']])
df_nominal_test = pd.DataFrame(index=x_a_test.index, data=nominal_test, columns=onehot.get_feature_names(['Facility_type_category', 'Owner_type', 'Regulatory_body']))
x_a_test = pd.concat([x_a_test, df_nominal_test], axis=1)
x_a_test.drop(columns=['Facility_type_category', 'Owner_type', 'Regulatory_body'], inplace=True)
x_a_test.head()
```

```python
#Feature Scaling (Data Standardization/Normalization)

from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler

ss = StandardScaler()
ss_train = ss.fit_transform(x_a_train)
x_a_train_ss = pd.DataFrame(ss_train)
x_a_train_ss.columns=x_a_train.columns
x_a_train_ss.head()

#Test
ss_test = ss.fit_transform(x_a_test)
x_a_test_ss = pd.DataFrame(ss_test)
x_a_test_ss.columns=x_a_test.columns
x_a_test_ss.head()
```

```python
#Oversampling (SMOTE: Synthetic Minority Oversampling Technique)

from imblearn.over_sampling import SMOTE
from collections import Counter

oversample = SMOTE()

x_a_train_ss_over, y_a_train_over = oversample.fit_resample(x_a_train_ss, y_a_train)
print(Counter(y_a_train_over))
```

```python
#Feature Importance

#5.1.1. Test on df_a set first
from sklearn.ensemble import ExtraTreesClassifier

fi = ExtraTreesClassifier()
fi_a = fi.fit(x_a_train_ss_over, y_a_train_over)

df_fi_a = pd.DataFrame(fi_a.feature_importances_,index=x_a_train_ss_over.columns)
df_fi_a.nlargest(10,df_fi_a.columns).plot(kind='barh')
plt.show()
```

```python
# Apply algorithm 
# Models to be tested: KNearest, SVM, Decision Tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
classifiers = {
    "KNeighbors" : KNeighborsClassifier(),
    "SVC" : SVC(),
    "DecisionTree" : DecisionTreeClassifier()
}
```

```python
#SVC

svc = SVC()
svc_params = {"C": [0.5, 0.7, 0.9, 1], "kernel":['rbf', 'poly', 'sigmoid', 'linear']}
grid_svc = GridSearchCV(svc, svc_params)
grid_svc.fit(x_a_train_ss_over, y_a_train_over)
svc_best = grid_svc.best_estimator_
```

```python
print('Accuracy of SVC classifier on this training is {:.2f}'
     .format(svc_best.score(x_a_train_ss_over, y_a_train_over)))
```

```python
#KNearest Neighbors
knear = KNeighborsClassifier()
knear_params = {"n_neighbors": list(range(2,7,1)), "algorithm": ['auto', 'ball_tree', 'kd_tree', 'brutle']}
grid_knear = GridSearchCV(knear, knear_params)
grid_knear.fit(x_a_train_ss_over, y_a_train_over)
knear_best = grid_knear.best_estimator_
```

```python
print('Accuracy KNeighbors classifier on this training is {:.2f}'
     .format(knear_best.score(x_a_train_ss_over, y_a_train_over)))
```

```python
#Decision Tree

tree = DecisionTreeClassifier()
tree_params = {"criterion": ['gini', 'entropy'], "max_depth":list(range(2,5,1)), "min_samples_leaf":list(range(5,7,1))}
grid_tree = GridSearchCV(tree, tree_params)
grid_tree.fit(x_a_train_ss_over, y_a_train_over)
tree_best = grid_tree.best_estimator_
```

```python
print('Accuracy of DecisionTree classifier on this training is {:.2f}'
     .format(tree_best.score(x_a_train_ss_over, y_a_train_over)))
```
